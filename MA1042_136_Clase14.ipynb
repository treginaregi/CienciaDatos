```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import scipy.stats as ss
import seaborn as sns

def atipicos(df,cols = None, method="IQR"):
  """ Se le entrega una data frame y devuelve la misma data frame pero sin los renglones que contengan algún atípico.
   Ver, por ejemplo, https://www.codegrepper.com/code-examples/python/remove+outlier+columns+pandas"""
  if cols==None:
    cols=df.columns
  ### Ver, por ejemplo, https://www.codegrepper.com/code-examples/python/remove+outlier+columns+pandas
  if method =="IQR":
    Q1=df[cols].quantile(0.25)
    Q3=df[cols].quantile(0.75)
    IQR = Q3- Q1
    mask = (df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))
    #print(mask)
    #print(mask.any(axis=1))
    atipicals = df[mask.any(axis=1)][cols]
    no_atipicals = df[~(mask).any(axis=1)][cols]                                               
    return atipicals, no_atipicals
  elif method=="mean-std":
    m=df[cols].mean()
    s=df[cols].std()
    mask = (df[cols] - m < -2*s) |(df[cols] - m > 2*s)
    #print(mask)
    #print(mask.any(axis=1))
    atipicals = df[mask.any(axis=1)][cols]
    no_atipicals = df[~(mask).any(axis=1)][cols]                                               
    return atipicals, no_atipicals
  else:
    print("Método no disponible")

df=pd.read_csv("Datos_buenos.csv",index_col=0)
print("********************************")
print("********************************")
print("Antes de quitar atípicos")
print("*********************************")
print("**********************************")
print("The columns are:")
print(list(df.columns))
print("Estadísitca descriptiva de los datos")
display(df.describe())
print("Columnas:")
print(df.columns)
print("******************")
print("*******************")
print("Descriptive plots")
plt.figure(figsize=(12,9))

n=1
for col in df.columns:
  y=df[col]
  order=range(len(y))

  #Histograma
  plt.subplot(3,5,n)
  plt.title(f"Histograma de {col}")
  plt.xlabel(f"{col}")
  plt.ylabel("Frecuencia")
  plt.hist(y)
  n=n+5

  #Boxplot
  plt.subplot(3,5,n)
  plt.title(f"Boxplot de {col}")
  plt.ylabel(f"{col}")
  plt.boxplot(y)
  n=n+5

  #Scatter
  plt.subplot(3,5,n)
  plt.title(f"Scatter de {col}")
  plt.xlabel("Order")
  plt.ylabel(f"{col}")
  plt.scatter(order,y)
  n=n-9
plt.tight_layout()
plt.show()
print("*******************")
print("*******************")
print("Modelo 1:")
# Modelo1: Con intercept y con Sodio
modelo1=smf.ols(formula="Cals ~ Sods + Fats + Carbs + Prots", data =df)
resultados1 = modelo1.fit()
print(resultados1.summary())

print("********************")
print("********************")
print("Modelo 2:")
# Modelo2: Sin intercept y con Sodio
modelo2=smf.ols(formula="Cals ~ 0 + Sods + Fats + Carbs + Prots", data =df)
resultados2 = modelo2.fit()
print(resultados2.summary())
print("*******************")
print("*******************")
print("Modelo 3:")
# Modelo3: Sin intercept y sin Sodio
modelo3=smf.ols(formula="Cals ~ 0 + Fats + Carbs + Prots", data =df)
resultados3 = modelo3.fit()
print(resultados3.summary())

print("****************")
print("*****************")
print("Gráfico 4-en-1")
residuals=resultados3.resid
fitted= resultados3.fittedvalues
order = range(len(residuals))

plt.figure(figsize=(6,6))
#Prueba de normalidad
plt.subplot(2,2,1)
plt.title("Prueba de Normalidad(Probplot)")
ss.probplot(residuals, plot= plt)

#Residuals Vs Fitted
plt.subplot(2,2,2)
plt.title("Residuals Vs Fitted")
plt.xlabel("fitted")
plt.ylabel("residuals")
plt.scatter(fitted, residuals)

#Histograma de residuals
plt.subplot(2,2,3)
plt.title("Histograma de residuals")
plt.xlabel("residuals")
plt.ylabel("frecuencia")
plt.hist(residuals)


#Residuals vs order
plt.subplot(2,2,4)
plt.title("Residuals vs Order")
plt.xlabel("Order")
plt.ylabel("Residuals")
plt.scatter(order, residuals)

plt.tight_layout()
plt.show()

df_with_res=df.copy()
df_with_res['residuals']=residuals

print("*****************************")
print("*****************************")
print("Atípicos en las variables usando IQR:")
display(atipicos(df)[0])
print("Atípicos en los residuos del modelo 3 usando IQR:")
display(atipicos(df_with_res,cols=['residuals'])[0])

df=atipicos(df_with_res)[1][['Sods', 'Fats', 'Carbs', 'Prots', 'Cals']]

print("********************************")
print("********************************")
print("Después de quitar atípicos")
print("*********************************")
print("**********************************")
print("The columns are:")
print(list(df.columns))
print("Estadísitca descriptiva de los datos")
display(df.describe())
print("Columnas:")
print(list(df.columns))
print("******************")
print("*******************")
print("Descriptive plots")
plt.figure(figsize=(12,9))

n=1
for col in df.columns:
  y=df[col]
  order=range(len(y))

  #Histograma
  plt.subplot(3,5,n)
  plt.title(f"Histograma de {col}")
  plt.xlabel(f"{col}")
  plt.ylabel("Frecuencia")
  plt.hist(y)
  n=n+5

  #Boxplot
  plt.subplot(3,5,n)
  plt.title(f"Boxplot de {col}")
  plt.ylabel(f"{col}")
  plt.boxplot(y)
  n=n+5

  #Scatter
  plt.subplot(3,5,n)
  plt.title(f"Scatter de {col}")
  plt.xlabel("Order")
  plt.ylabel(f"{col}")
  plt.scatter(order,y)
  n=n-9
plt.tight_layout()
plt.show()
print("*******************")
print("*******************")
print("Modelo 1:")
# Modelo1: Con intercept y con Sodio
modelo1=smf.ols(formula="Cals ~ Sods + Fats + Carbs + Prots", data =df)
resultados1 = modelo1.fit()
print(resultados1.summary())

print("********************")
print("********************")
print("Modelo 2:")
# Modelo2: Sin intercept y con Sodio
modelo2=smf.ols(formula="Cals ~ 0 + Sods + Fats + Carbs + Prots", data =df)
resultados2 = modelo2.fit()
print(resultados2.summary())
print("*******************")
print("*******************")
print("Modelo 3:")
# Modelo3: Sin intercept y sin Sodio
modelo3=smf.ols(formula="Cals ~ 0 + Fats + Carbs + Prots", data =df)
resultados3 = modelo3.fit()
print(resultados3.summary())

print("****************")
print("*****************")
print("Gráfico 4-en-1")
residuals=resultados3.resid
fitted= resultados3.fittedvalues
order = range(len(residuals))

plt.figure(figsize=(6,6))
#Prueba de normalidad
plt.subplot(2,2,1)
plt.title("Prueba de Normalidad(Probplot)")
ss.probplot(residuals, plot= plt)

#Residuals Vs Fitted
plt.subplot(2,2,2)
plt.title("Residuals Vs Fitted")
plt.xlabel("fitted")
plt.ylabel("residuals")
plt.scatter(fitted, residuals)

#Histograma de residuals
plt.subplot(2,2,3)
plt.title("Histograma de residuals")
plt.xlabel("residuals")
plt.ylabel("frecuencia")
plt.hist(residuals)


#Residuals vs order
plt.subplot(2,2,4)
plt.title("Residuals vs Order")
plt.xlabel("Order")
plt.ylabel("Residuals")
plt.scatter(order, residuals)

plt.tight_layout()
plt.show()
print("************************")
print("*************************")
print("Media de los residuos del modelo3:")
print(residuals.mean())
print("Desviación estándard de los residuos del modelo 3:")
print(residuals.std())

print("*********************************************")
print("*********************************************")
print("Correlaciones en el modelo final")
display(df.corr())
sns.pairplot(data=df)
plt.show()
```
